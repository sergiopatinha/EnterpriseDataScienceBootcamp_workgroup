# 1: Dataset Aquisition

1. Importing the libraries/functions:

```python
import pandas as pd
import openpyxl
import os, re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

2. Paths and names of the files

```python
DATA_DIR = os.path.join("..", "data/raw")

FILES = {
    "demographics": "Telco_customer_churn_demographics.xlsx",
    "location":     "Telco_customer_churn_location.xlsx",
    "population":   "Telco_customer_churn_population.xlsx",
    "services":     "Telco_customer_churn_services.xlsx",
    "status":       "Telco_customer_churn_status.xlsx",
}

dfs = {name: pd.read_excel(os.path.join(DATA_DIR, fname)) for name, fname in FILES.items()}
```

```python
demographics = dfs["demographics"]
location = dfs["location"]
population = dfs["population"]
services = dfs["services"]
status = dfs["status"]
```

3. Read of the files and normalization of 'CostumerID'

```python
dfs = {}

for name, fname in FILES.items():
    path = os.path.join(DATA_DIR, fname)
    df = pd.read_excel(path)

    ##normalize column names
    df.columns = [re.sub(r"\s+", " ", c).strip() for c in df.columns]

    # normalize CustomerID
    if "CustomerID" in df.columns:
        df["CustomerID"] = df["CustomerID"].astype(str).str.strip()

    dfs[name] = df
    print(f"{name}: {df.shape}")
```

It was also identified the possibility of having a union on the datasets based on the Customer ID. Let's do it, step by step:
1. cleaning blank spaces on the column names (to avoid duplicating collumns)
2. check which columns collide before merging them
3. normalize the customer ID
4. merge the tables



```python
##Clean blank spaces on the column names

for df in [demographics, location, services, status]:
    df.columns = (
        df.columns
        .str.strip()               #remove spaces at the beginning and end
        .str.replace(r"\s+", "_", regex=True)  #replace spaces with underscores
        .str.lower()                # transform to lowercase
    )

print(df)
```

```python
#check which columns collide before merging them

def overlapping_cols(a, b, key="customer_id"):
    return sorted((set(a.columns) & set(b.columns)) - {key})

print("demo vs loc:", overlapping_cols(demographics, location))
print("demo+loc vs svc:", overlapping_cols(
      demographics.merge(location, on="customer_id", how="inner"), services))
print("… vs status:", overlapping_cols(
      demographics.merge(location, on="customer_id", how="inner")
                  .merge(services, on="customer_id", how="inner"), status))
```

```python
#normalize the Customer ID

KEY = "customer_id"  # already normalized to lower/underscore

def add_prefix_except(df: pd.DataFrame, prefix: str, keep=(KEY,)) -> pd.DataFrame:
    return df.rename(columns={c: (prefix + c) if c not in keep else c for c in df.columns})

# normalize function to standardize column names and customer_id
def normalize(df):
    out = df.copy()
    out.columns = (out.columns
                   .str.strip()
                   .str.replace(r"\s+", "_", regex=True)
                   .str.lower())
    for cand in ("customer_id","customerid","customer_id_","customer_id1","customer id"):
        if cand in out.columns and cand != KEY:
            out = out.rename(columns={cand: KEY})
    out[KEY] = out[KEY].astype(str).str.strip()
    return out

demographics = normalize(demographics)
location     = normalize(location)
services     = normalize(services)
status       = normalize(status)

# prefixes to avoid collisions
demo_ = add_prefix_except(demographics, "demo_")
loc_  = add_prefix_except(location,     "loc_")
svc_  = add_prefix_except(services,     "svc_")
st_   = add_prefix_except(status,       "st_")

# safe merge (one_to_one guarantees no duplicate keys)
df_master = (demo_
             .merge(loc_, on=KEY, how="inner", validate="one_to_one")
             .merge(svc_, on=KEY, how="inner", validate="one_to_one")
             .merge(st_,  on=KEY, how="inner", validate="one_to_one")
            ).set_index(KEY)

print(df_master.shape)
df_master.head()
df_master.info()
```
